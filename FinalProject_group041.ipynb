```python
#function to remove the percent signs
def rem_perc(string):
    
    string = string.lower()
    string = string.strip()

    string = string.replace('%', '')

    # restrip the output
    string = string.strip()
    
    output = float(string);
    
    return output
```

```python
#remove the percent signs from 'GDP growth' series
bangladesh_econ['GDP growth'] = bangladesh_econ['GDP growth'].apply(rem_perc);
```

```python
#remove the percent signs from 'Inflation rate' series
bangladesh_econ['Inflation rate'] = bangladesh_econ['Inflation rate'].apply(rem_perc);
```

```python
#remove the percent signs from 'Total Investment' series
bangladesh_econ['Total Investment'] = bangladesh_econ['Total Investment'].apply(rem_perc);
```

```python
#function to remove the commas
def rem_commas(string):
    
    string = string.lower()
    string = string.strip()

    string = string.replace(',', '')

    # restrip the output
    string = string.strip()
    
    output = int(string);
    
    return output
```

```python
#remove the commas from 'GDP per capita' series
bangladesh_econ['GDP per capita'] = bangladesh_econ['GDP per capita'].apply(rem_commas);
```

## Inferential Analysis - Population Growth and AQI


The next section of the analysis visualizes the population growth and AQI data together for the countries. By putting the data together, we are able to find relationships between the data and see if there are any correlations visible. 

The analysis and association between the variables can be found in the next section with the r and p values calculated.

- **Null Hypothesis:** The population growth rate of a country has no direct correlation to the country’s AQI score.
- **Alternative Hypothesis:** There is a direct correlation between a country’s population growth rate and their AQI. 
- **Alpha/Significance Level:** 0.05
- **Test Statistic:** Pearson (r) Test

Since our dataset contains quantitative data, we can use the Pearson (r) test.

Through the use of a Pearson (r) test, we can determine if there is a relationship between the population growth rate and AQI levels for each country. The AQI levels are divided into minimum and maximum exposure values for each year, so we will be performing Pearson (r) tests on both values provided.

r test indicates that the relationship between the variables is relatively strong. R is negative (-0.829) which means that there is an inverse correlation between the variables. The p value (0.009) is less than 0.05 and thus we reject the null hypothesis. The alternative hypothesis is true, which statest that there is a direct correlation between a country’s population growth rate and their AQI.

## Inferential Analysis - Poverty and AQI

- __Null Hypothesis__: Poverty rate is not associated with minimum/maximum AQI levels. 
- __Alternative Hypothesis__: Poverty rate is associated with minimum/maximum AQI levels. 
- Alpha/Significance Level: 0.05
- Test Statistic: Pearson (r) Test 

We chose to use a Pearson (r) test to determine if there is a relationship between the quantitative variables poverty rate and AQI levels for each country. The AQI levels are divided into minimum and maximum exposure values for each year, so we will be performing Pearson (r) tests on both values provided. Since our dataset contains quantitative data, we can use the Pearson (r) test. 
- All null hypotheses are in the form of: "The distribution of poverty rate is the same for X and Y", with each country as the different variable.

## Inferential Analysis - GDP and AQI

__Null Hypothesis:__ GDP rate is not associated with minimum/maximum AQI levels.

__Alternative Hypothesis:__ GDP rate is associated with minimum/maximum AQI levels.

__Alpha/Significance Level:__ 0.05

__Test Statistic:__ Pearson (r) Test

We chose to use a Pearson (r) test to determine if there is a relationship between the quantitative variables GDP rate and AQI levels for each country. The AQI levels are divided into minimum and maximum exposure values for each year, so we will be performing Pearson (r) tests on both values provided. Since our dataset contains quantitative data, we can use the Pearson (r) test.
All null hypotheses are in the form of: "The distribution of inflation rate is the same for X and Y", with each country as the different variable.

### Bangladesh r Test

```python
stats.pearsonr(BGD_GDP['Exposure Lower'], BGD_GDP['Annual GDP Rate'])
#r value = 0.632
#p-value = 0.028
```

Due to the small p value (0.028 < 0.05), we reject the null hypothesis. GDP rate might be associated with minimim AQI levels. There is also a strong positive correlation, as determined by the large positive r value of 0.632.


```python
stats.pearsonr(BGD_GDP['Exposure Upper'], BGD_GDP['Annual GDP Rate'])
#r value = -0.252
#p-value = 0.431
```

Due to the large p value (0.431 > 0.05), we fail to reject the null hypothesis. GDP rate is not associated with maximum AQI levels. There is also a small negative r value of -0.252. This number suggests that the data is slightly negatively correlated, but not strongly for a correlation to be observed.

Data manipulation can cause limitations in the datasets as well as give an inaccurate picture of the figures. A good way to test against data manipulation is through the use of the chi-square goodness of fit test. This statistical hypothesis can be used to determine how well a theoretical normal distribution can fit the empirical distribution. By dividing the sample data into intervals, we can compare the number of points that have fallen into the interval against the expected number of points in each interval. Moreover, we can also use Benford’s Law which is capable of predicting alteration of numbers in a set of data. This statistical technique can be used to show differences between what is shown and what the data would be if naturally collected.

If the sample size did not accurately reflect in the study or if it was insufficient, we would avoid these datasets out of caution.
